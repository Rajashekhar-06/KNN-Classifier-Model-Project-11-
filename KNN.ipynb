```python
my_data=load_iris()
my_data
```

```python
print(my_data.DESCR)
```

```python
feature=pd.DataFrame(my_data.data,columns=my_data.feature_names)
feature
```

```python
target=pd.DataFrame(my_data.target,columns=["target"])
target
```
